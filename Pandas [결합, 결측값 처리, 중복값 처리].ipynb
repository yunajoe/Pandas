```python
# Combining data
#  CONCAT 

import pandas as pd 
import numpy as np

df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                   'b':['b0','b1','b2','b3'],
                   'c':['c0','c1','c2','c3']},
                  index = [0,1,2,3])  # 4 X 3 metrix


df2 = df2 = pd.DataFrame({'a':['a0','y','z','h'],
                   'b':['b0','b1','b2','b3'],
                   'c':['c0','y2','z2','h2'],
                   'd':['d0','y3','z3','h3']},
                   index = ["영","일","이","삼"]) # 4  X 4 metrix

print(df1, "\n")
print(df2)
```

```python
result1 = pd.concat([df1,df2], axis= 0)  # 행 방향으로 이어붙이기 
print(result1)
```

```python
result2 = pd.concat([df1,df2],ignore_index = True)  #  인덱스를 정리해어 붙이기
print(result2)
```

```python
result3 = pd.concat([df1,df2],axis=1) # 열방향으로 가져오기 
print(result3)
```

```python
# merge는 공통의 열을 기준으로 두 데이터 프레임을 합쳐준다 

# 기준열 이름이 같을 때 
# pd.merge(left, right, on = '기준열', how = '조인방식')

# 기준열 이름이 다를 때
# pd.merge(left, right, left_on = '왼쪽 열', right_on = '오른쪽 열', how = '조인방식')

# 조인방식 (left, right, inner, outer)  이 있고 inner가 default 

```

```python
Fruit = pd.DataFrame({"Number":[1,2,3,4,5],"Fruit":['Apple', 'Banana', 'Cherry', 'Lemon', 'Peach']})

Grade = pd.DataFrame({"Number":[1,2,3], "Grade":["A","B","C"]})

print(Fruit,"\n")
print(Grade)

# 두 개의 df 는 Number 라는 열을 공통으로 가진다. 
```

```python
# Left Join  왼쪽 데이터프레임을 기준으로  조인하기

# pd.merge(left, right, on = "기준열", how = "left")

pd.merge(Fruit, Grade, on = "Number", how = "left")
```

```python
# Right join 오른쪽 데이터프레임을 기준으로 조인하기 
# pd.merge(left, right, on = "기준열", how = "right")

pd.merge(Fruit,Grade,on = "Number",how = "right")
```

```python
# inner join  교집합으로 양쪽에 공통으로 있는 값만 나타낸다
# pd.merge(left, right, on = '기준열', how = 'inner')
pd.merge(Fruit,Grade, on = "Number", how = "inner")
```

```python
# outer join 합집합 
# pd.merge(left, right, on = "기준열", how = "outer")
pd.merge(Fruit, Grade, on = "Number", how = "outer")
```

```python
# 공통이 되는 기준열이 여러개 일 때 
# pd.merge(left, right, on = [기준열1,기준열2...])

Fruit = pd.DataFrame({"Number":[1,2,3,4,5],"Fruit":['Apple', 'Banana', 'Cherry', 'Lemon', 'Peach']})

Grade = pd.DataFrame({"Number":[1,2,3], "Fruit":['Apple', 'Banana', 'Cherry'],"Grade":["A","B","C"]})

print(Fruit, "\n")
print(Grade)
```

```python
left_merge = pd.merge(Fruit, Grade, on= ["Number", "Fruit"], how = "left")
right_merge = pd.merge(Fruit, Grade, on =["Number", "Fruit"], how = "right")
inner_merge = pd.merge(Fruit, Grade, on =["Number", "Fruit"], how = "inner")
outer_merge = pd.merge(Fruit, Grade, on =["Number", "Fruit"], how = "outer")

print(left_merge)
```

```python
print(right_merge)
```

```python
# 두 df의 열 이름이 다를 때 
# pd.merge(left, right, left_on = '왼쪽 열', right_on = '오른쪽 열')



Fruit = pd.DataFrame({"Number":[1,2,3,4,5],"Fruit":['Apple', 'Banana', 'Cherry', 'Lemon', 'Peach']})

Grade = pd.DataFrame({"순서":[1,2,3], "Grade":["A","B","C"], "Price":[100,200,300]})
```

```python
pd.merge(Fruit, Grade, left_on = "Fruit",right_on= "Grade", how="left")
```

```python
pd.merge(Fruit, Grade, left_on = "Fruit",right_on= "Grade", how="right")
```

```python
# 결측치(NaN), 중복데이터 처리하기 

import seaborn as sns 
import pandas as pd 
import numpy as np 

df = pd.read_csv("C:\\Users\\yuna\Downloads\\titanic\\test.csv")
df.head()
```

```python
df.info() # 데이터를 간단히 살펴보기 
# 총 11개의 columns , 418개의 rows, "AGE", "CABIN","FARE" 가 결측치(NaN)을 가지고 있다.
```

```python
print(df["Age"].value_counts,"\n")
print("dropna를 추가했을때")
print(df["Age"].value_counts(dropna= False))  # Dropna = False 를 하면 Nan 값을 추가해서 카운트해줌
```

```python
# isnull()  & notnull() 을 사용하면은 결측치(NaN)에 대해서 true/false 를 돌려줌 

df.head().isnull()  # NaN일 때 True 반환
```

```python
df.head().notnull() # NaN일 때 False 반환
```

```python
# dropna 를 통해 결측치를 제거해보자 

df.dropna(axis=0).info()  # NaN 이 있는 행은 제거해서 87개 만 줄어든 것을 볼 수 있음
```

```python
df.dropna(axis=1).info()  # axis = 1을 하면 NaN 이 존재하는 열은 삭제해 버려서 컬럼수가 11개에서 8개로 줄어듬
```

```python
# thresh = xx   ( 결측치가 xx 이상인 행 또는 열을 모두 삭제해 버림)

df_thresh = df.dropna(axis = 1, thresh = 100)  # 결측치가 100개 이상인 열이었던 Cabin 삭제
df_thresh.info()

```

```python
# subset = ["열"] 옵션 
# 열 옵션에 결측값이 1개라도 있으면 그 행을 drop시키자 
df_age = df.dropna(subset = ["Age"], how = "any", axis = 0)
# 만약 how 가 all 이면은 NaN이 모든 열 또는 행에 존재해야함 

print(len(df_age),"\n") # NaN이 없는, 값만 들어있는 행들 
print(df_age["Age"].isnull().sum())
```

```python
import seaborn as sns 

df = pd.read_csv("C:\\Users\\yuna\Downloads\\titanic\\test.csv")
```

```python
print("*** 변경 전 ***")
print(df["Age"].head(30))
mean_age = df["Age"].mean()   # Age의 평균값을 만들기 
df["Age"].fillna(mean_age,inplace = True)# 평균값으로 대체하기 
print("\n")
print("*** 평균 값 ***")
print(mean_age)
print("\n")
print("*** 변경 후 *** ")
print(df["Age"].head(30))
```

```python
print(df["Fare"].value_counts())
print('\n')
print("***결측치갯수***")
print(df["Fare"].isnull().sum()) 


```

```python
# 결측치 1개가 어느 index, row에 있는지 확인해보기 

df.index[df["Fare"].isnull()]   # 152행에 NaN이 있다고 나와 있음 
```

```python
# 확인해보자 
df.iloc[152]
```

```python
# Fare 의 NaN 값을 최빈값으로 대체해보자 
# idxmax() 는 최대갯수 값을 가지는 index를 불러와줌
most_freq = df["Fare"].value_counts(dropna= True).idxmax()  # dropna = False를 하면은 NaN값도 같이 불러오기 때문에 , dropna = True인 default 값을 불러오자
most_freq  # 최빈값은  7.75
```

```python
df["Fare"].fillna(most_freq, inplace = True)
df.iloc[152] # NaN값이었던 Fare가 7.75로 바뀐것을 볼 수 있음
```

```python
# NaN의 값을 이웃값으로 대체해 보기 
# method = "ffill" 은 NaN의 바로의 값으로, method = "bfill"의 NaN의 뒤의 값으로 변경
import seaborn as sns 
df = pd.read_csv("C:\\Users\\yuna\Downloads\\titanic\\test.csv")

print("*** 변경 전 ***")
print(df["Fare"][150:155])
print("\n")

print("*** NaN 이전 값으로 변경하기 *** ")
df["Fare"].fillna(method = "ffill", inplace = True)
print(df["Fare"][150:155])
```

```python
# 중복 데이터 
# 중복 데이터 탐색 

import pandas as pd 
df = pd.DataFrame({'c1' : ['a','a', 'b', 'a', 'b'],
                  'c2' : [1, 1, 1, 2, 2],
                   'c3' : [1, 1, 2, 2, 2]})

df
```

```python
# duplicated()
# 전에 나온 행들과 비교하여 중복되는 행이면 True, 아니면 False
# 첫 번째(0번째) 값은 항상 False
# 바로 전 행과만 비교하는게 아니라 처음 나오는 행인지, 나왔던 행인지를 비교
# 즉, False의 개수가 유니크한 값의 개수

df.duplicated()
```

```python
col_2 = df["c2"].duplicated()
col_2
```

```python
# Series에서도 적용이 가능 
test = pd.Series([3,1,2,1,2,2,3,])
test.duplicated()
```

```python
# 중복 데이터 제거 
# drop_duplicates()

import pandas as pd 

df = pd.DataFrame({'c1' : ['a','a', 'b', 'a', 'b'],
                  'c2' : [1, 1, 1, 2, 2],
                   'c3' : [1, 1, 2, 2, 2]})

df
```

```python
# duplicated()는 중복행을 탐색 하는 작업이라면, 
# drop_duplicates() 중복행을 제거해 주는 역할 

df2 = df.drop_duplicates()
df2
```

```python
# subset = [열1, 열2..] 특정열을 기준으로 중복행 제거 
print(df)
print("\n")
print("*** 콜럼 c2, c3를 기준으로 중복 열 제거해주기 *** ")

df3 = df.drop_duplicates(subset = ["c2","c3"])
print(df3) # 행 1,4행이 0, 3행과 중복되었다는 것을 알 수 있음
```
