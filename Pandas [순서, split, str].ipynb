```python
import seaborn as sns 
titanic = sns.load_dataset("titanic")
df = titanic.loc[0:4,"survived":"age"]
df
```

```python
col1 = list(df.columns)
print(col1,"\n")

print(" sorted (오름차순)으로 정렬한 결과")
col2 = sorted(df.columns) # sorted 는 오름차순으로 정렬을 해 준다 
print(col2, "\n")
```

```python
df2 = df[col2]
df2   # 오름차순으로 콜럼들을 정렬한 결과 
```

```python
# reversed 를 이용해서 역순으로 정렬해보자 
reversed_col = list(reversed(df.columns))
print(reversed_col,"\n")
df3 = df[reversed_col]
df3
```

```python
# 열분리하기 

import pandas as pd 

df = pd.read_excel("C:\\Users\\yuna\\Downloads\\주가데이터.xlsx")
print(df)
print("\n")
print(df.dtypes)
```

```python
# 연월일 컬럼을 분리해보자 

df["연월일"] = df["연월일"].astype("str")
print(df.dtypes) # 연월일 컬럼이 datetime 에서 str 로 변경된 것을 확인할 수있다. 

```

```python
dates = df["연월일"].str.split("-")
dates
```

```python
df["연"] = dates.str.get(0)   # index 0 인 년만 뽑아내기 
df["월"] = dates.str.get(1)   # index 1인 월만 뽑아내기
df["일"] = dates.str.get(2)   # index 2인 일만 뽑아내기 
df
```

```python
# mapping &  for 문을 이용해 같은 결과를 출력해보자 

df = pd.read_excel("C:\\Users\\yuna\\Downloads\\주가데이터.xlsx")
df
```

```python
df["연월일"] = df["연월일"].astype("str")   # str로 데이터 타입 변경해주기 

col = ["연","월","일"]
for i in range(len(col)):
    df[col[i]] = df["연월일"].apply(lambda x: x.split("-")[i])
    
df.head()
```

```python
# 시자글자 인식 
#  .str.startswith()
df = pd.read_excel("C:\\Users\\yuna\\Downloads\\법정동.xlsx")
df
```

```python
# 끝글자 인식 
# .str.endswith()
# 동으로 끝나는 단어를 찾아보자
df["법정동명"].str.endswith("동")
```

```python
# 동으로 끝나는 데이터만 필터링해보자
df[df["법정동명"].str.endswith("동")]
```

```python
# 포함글자 인식 
# .str.contains()
# 강서구가 들어간 데이터를 찾아보자
df = pd.read_excel("C:\\Users\\yuna\\Downloads\\법정동.xlsx")
df
```

```python
df["법정동명"].str.contains("강서구")
# 강서구가 들어간 데이터는 True,아니면 False 로 반환 
```

```python
# 강서구가 들어간 데이터만 필터링
df[df["법정동명"].str.contains("강서구")]
```

```python
# 문자위치 찾기 
# 왼쪽부터 검색 후, 찾으려는 문자가 없으면 -1
print(df["법정동명"])
print("\n")
print(df["법정동명"].str.find("동"))
```

```python
# 오른쪽부터 검색 후, 찾으려는 문자가 없으면 -1  ( rfind)
df["법정동명"].str.rfind("강")
```

```python
# 찾으려는 값 모두 반환 (findall)
df["법정동명"].str.findall("\w+동")
```

```python
# 문자 대체 ( replace)
df["법정동명"].str.replace(" ","**")  # 공백으로 ** 로 대체하자
```

```python
# 원하는 문자열 추출 ( str.extract())
df["법정동명"].str.extract(r"(\w*동)")

#패턴에 맞는 단어가 없을 시 NaN 이 출력
```

```python
# 문자열 padding 
# 문자열 길이가 고정되어 부족할 경우 채워야 할 떄가 있음
#  .str.pad 

```

```python
# 문자열 길이 20자, 왼쪽부터"-" 로 채우기
df["법정동명"].str.pad(width = 20, side ="left", fillchar="_")
```

```python
# 문자열 길이 20자, 오른쪾부터"-" 로 채우기
df["법정동명"].str.pad(width = 20, side = "right",fillchar = "_")
```

```python
# 문자열 길이 20자, 좌우로 "_"로 채우기
df["법정동명"].str.center(width = 20,fillchar = "_")
```

```python
# 왼쪽부터 0으로 채우기 
df["법정동명"].str.zfill(width = 20)
```
