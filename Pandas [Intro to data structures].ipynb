```python
# Series ( 일차원 데이터로써 정수, 문자, 실수, python object가 될 수 있음)
# 형태 s = pd.Series(data, index=index)

import numpy as np
import pandas as pd 

# From radom.randn 

s = pd.Series(np.random.randn(5), index =["a","b","c","d","e"])
print(s)
print("\n")
print(s.index)

# From dictionary 
print("\n")
d = {"a":0,"b":1,"c":2}
d1 = pd.Series(d)
print(d1)
print("\n")
d2 = pd.Series(d,index = ["a","c","b","d"])
print(d2)
```

```python
# From scalar value
pd.Series(5, index= ["a","b","c","d","e"])
```

```python
s = pd.Series(np.random.randn(5), index =["a","b","c","d","e"])
print(s+s)
print("\n")
print(s*2)
print("\n")
print(s[1:] + s[:1])
```

```python
# Name attribute

s = pd.Series(np.random.randn(5), name = "something")
s
```

```python
# rename a Series with the pandas.Series.rename()

s2 = s.rename("different")
print(s2)
print("\n")
print(s2.name)
```

```python
# DafaFrame (2차원 배열의 데이터)
# From dict of ndarrays / lists

import pandas as pd 
import numpy as np

d= {"one":[1,2,3,4], "two":[5,6,7,8]}
df = pd.DataFrame(d)
print(df)
print("\n")
df2 = pd.DataFrame(d, index = ["a","b","c","d"])  # index 변경하기 
print(df2)
```

```python
# From a list of dicts

data = [{"a":1,"b":2},{"a":5,"b":10,"c":20}]

data1 = pd.DataFrame(data)
print(data1)
print("\n")
data2 = pd.DataFrame(data,index = ["first","second"]) # index 변경하기
print(data2)
print("\n")
data3 = pd.DataFrame(data,columns = ["a","b"]) # a, b 컬럼만 갖구오기
print(data3)
```

```python
# Column selection, addition, deletion

df = pd.DataFrame({"one":[1,2,3],"two":[4,5,6],"three":[7,8,9]}, index = ["A","B","C"])
df
```

```python
df["one"]
```

```python
df["multiply"] = df["one"] * df["two"]
df
```

```python
df["boolen"] = df["one"]> 2
df
```

```python
# deleted or popped

del df["two"]   # del 를 이용해 지우기 
df
```

```python
df.pop("three")  # pop을 이용해 지우기 
df
```

```python
df["foo"] = "bar"
df
```

```python
df["one_trunc"] = df["one"][0:1]  # df의 0 index(1행) 까지만 가져온다
df
```

```python
# insert 로 원하는 위치에다 컬럼 추가하기 

df.insert(0,"one_copy",df["one"])
df
```
