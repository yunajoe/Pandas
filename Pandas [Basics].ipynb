```python
#판다스는 2가지 오브젝트 Series, DataFrame이 있다. 

1) Series: 1차원 데이터와 인덱스로 구성(하나의 컬럼)
2) Dataframe : 2차원 데이터와 인덱스로 구성(여러개의 컬럼)
```

```python
# dataframe 만들기 

data = {"이름":["연아","민영","수민"],"나이":[10,12,12],"취미":["음악감상","운동","독서"]}
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data, index = ["일","이","삼"]) # row 이름 붙이깅
df
```

```python
# 행 & 이름 변경하기 
# DataFrame 객체.index = []
# DataFrame 객체.columns = []
df.index = ["첫번째","둘째","셋째"]
df.columns = ["Name","Age","Hobby"]
df
```

```python
#  행 인덱스 & 열 이름 변경 (2) 
# DataFrame 객체.rename(index={기존 인덱스:새 인덱스,---})
# DataFrame 객체.rename(columns={기존 이름:새 이름, ---})
df.rename(columns = {"Name":"이름","Age":"나이","Hobby":"취미"}, inplace = True)
df.rename(index = {"첫번째":"1st", "둘째":"2nd", "셋째":"3rd"}, inplace = True)
df
```

```python
# 행  & 열 삭제 
# 행 삭제 : DataFrame 객체.drop('행 인덱스' 또는 ['배열'], axis=0)
# 열 삭제 : DataFrame 객체.drop('열 이름' 또는 ['배열'], axis=1)

df.drop("2nd", axis = 0, inplace = True ) # 행 이름이 "2nd" 를 삭제하자 
df
```

```python
df.drop("나이", axis = 1, inplace = True)  # 열 이름이 나이를 삭제하자 
df
```

```python
# 데이터프레임 만들기(2)

exam = pd.DataFrame({"수학":[100,20,30],"영어":[90,39,30],"국어":[80,50,30] , "체육":[60,50,40]}
                   , index = ["연아","연아2","연아3"]) 

exam
```

```python
# 특정 행 선택하기 
# .loc 인덱스 이름으로 
# .iloc 인덱스 위치로 

index1 = exam.loc["연아"]
index2 = exam.iloc[0:2]
print(index1)  
print("\n")
print(index2)
```

```python
math = exam["수학"]
print(math)
```

```python
Eng = exam.영어
print(Eng)
```

```python
subject = exam[["수학","체육"]]
subject
```

```python
# 행/ 열 추가 
# 행추가 : DataFrame.loc['새로운 행 이름'] = 데이터 값 (또는 배열)
exam.loc["연아4"] = [10,20,30,40]
exam.loc["연아5"] = 1 
exam.loc["연아6"] = exam.loc["연아5"]
exam
```

```python
# 열 추가하기
# 열 추가 : DataFrame 객체[ '추가하려는 열 이름' ] = 데이터 값

exam["음악"] = 10,20,30,40,50,60
exam
```

```python
# 행, 열 함께 선택하기 
# 인덱스 이름 : DataFrame 객체.loc[행 이름, 열 이름]
# 정수 위치 인덱스 : DataFrame 객체.iloc[행 번호, 열 번호]
# 원소 1개 추출
a = exam.loc["연아", "체육"]
a
```

```python
exam.loc["연아",["수학", "영어"]] = 70,60 
exam # 연아 행의 수학, 영어 점수가 70,60으로 바뀐것을 알 수 있다.
```

```python
exam.iloc[0, [1,2]] = 11, 22  # 0행, 즉 연아 행의 1,2 컬럼(영어, 국어)의 점수가 11, 22으로 바뀜
exam
```

```python
exam.transpose()  # 행, 열 위치 바꾸깅
```

```python
# 인덱스 활용
# 1) .set_index  특정 열을 행 인덱스로 정한다 
# 특정 열을 행 인덱스로 설정 : DataFrame 객체.set_index( ['열 이름' ] 또는 '열 이름')
# 원본 데이터 프레임을 바꾸지 않고 새로운 데이터 프레임을 반환
```

```python
# 수학열을 인덱스로 설정하자 
df = exam.set_index(["수학"])
df
```

```python
df2 = exam.set_index(["수학","영어"])
df2
```

```python
# 행 인덱스 (.reindex)
# 새로운 배열로 행 인덱스를 재지정 : DataFrame 객체.reindex(새로운 인덱스 배열)

new_index = ["연아","연아2","연아3","yuna","yuna2","yuna3"]
exam.reindex(new_index)
# 기존에 존재않는 행 인덱스가 새로 생길 경우 값은 NaN 가 된다
```

```python
# 3) 행 인덱스 초기화 reset_index()
exam.reset_index()
```

```python
# 4) 행 인덱스를기준으로 데이터 프레임 정렬 .sort_index()
# ascending=True or False 옵션을 사용해 오름차순 또는 내림차순을 설정

a = exam.sort_index(ascending = False) # 내림차순 
b = exam.sort_index(ascending = True) # 오름차순
print(a)
print(b)
```

```python
print(exam.sort_values(by = "수학"))  # 수학점수 오름차순
# 특정 열(수학)를 기준으로 데이터 값을 기준으로 정렬하기 
print("\n")
print(exam.sort_values(by = "수학", ascending = False)) # 수학점수 내림차순
```
