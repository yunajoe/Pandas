```python
import pandas as pd
df = pd.read_csv("C:\\Users\\yuna\\Downloads\\laptops.csv")
df.head()
```

```python
brand_nation = {"Dell":"US", "Apple":"US","Acer":"Taiwan",
                "HP":"US","Lenova":"China","Microsoft":"US", 
                "Asus":"Taiwan","Alienware":"US"}
```

```python
df["brand"]
print("\n")
BrandDf = df["brand"].map(brand_nation)  # map의 기능: brand_nation 의 왼쪽의 있는 값을 오른쪽으로 바꿈 
BrandDf.head(20)
```

```python
df["Brand_nation"] = df["brand"].map(brand_nation )
df
```

```python
# groupby 
nation_groups = df.groupby("Brand_nation")
type(nation_groups)
```

```python
# 국가별로 종류를 나타냄
nation_groups.count()
```

```python
# 브랜드별 평균 가격을 구해보자 
brand_group = df.groupby("brand")
brand_group
```

```python
brand_price = brand_group.mean().sort_values(by = "price", ascending= False)
# sort_values() 경우에는 어떤 컬럼의 값을 기준으로 정렬해야 하는지 명시적으로 설정해야 하기때문에 "by" 를 넎는다
# price 를 기준으로 정렬한다 
brand_price["price"]
```

```python
# 데이터 합치기 
import pandas as pd
price_df = pd.read_csv("C:\\Users\\yuna\\Downloads\\vegetable_price.csv")
quantity_df = pd.read_csv("C:\\Users\\yuna\\Downloads\\vegetable_quantity.csv")
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df,quantity_df, on="Product") # inner join 과 같은역할
```

```python
pd.merge(price_df, quantity_df,on = "Product",how = "left")  # left join 과 같은 역할
```

```python
pd.merge(price_df,quantity_df,on="Product", how = "right") # right join과 같은 역할
```

```python
pd.merge(price_df, quantity_df, on ="Product", how= "outer") # 모든컬럼이 다 들어가있는 거
```
