```python
## 연습문제 

df = sns.load_dataset("titanic")
df.head()
```

```python
# 나이가 30살 이상 남자 승객 조건 필터링
# fare를 많이 낸 순서로 내림차순 정렬
# 상위 10개를 출력
```

```python
c1 = df["age"] >= 30 
c2 = df["sex"] == "male"
answer1 = df.loc[c1 & c2].sort_values(by="fare", ascending = False)
answer1.head(10)
```

```python
# 나이가 20살 이상 40살 미만인 승객
# pclass가 1등급 혹은 2등급인 승객
# 열(column)은 survived, pclass, age, fare 만 나오게 출력
# 10개만 출력
```

```python
df
```

```python
c1 = (df["age"] >=20) & (df["age"] < 40)
c2 = (df["pclass"] == 1)|(df["pclass"] == 2)
answer2 = df.loc[c1 & c2,["survived", "pclass", "age", "fare"]]
answer2.head(10)

# df[행 인덱스 값, 열 인덱싱 값]
```

```python
# pclass가 1, 2등급인 승객 중 나이 값이 없는(NaN) 데이터는
# 나이의 평균 값으로 채워라 

c1 = (df["pclass"] ==1)|(df["pclass"] ==2)

answer3 = df.loc[c1]
answer3
```

```python
answer3["age"].isnull().sum() # 총 41개 null값이다. age 컬럼 중에서 
```

```python
mean_age = answer3["age"].mean()
mean_age
```

```python
answer3["age"] = answer3["age"].replace(to_replace = np.nan, value= mean_age)
answer3["age"]
```
