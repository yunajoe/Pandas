```python
# apply, map, applymap 함수 

# 1) apply 는 Series 나 Dataframe 의 객체 요소에 적용시킬 수 있음
# 2) map 은 Series에만 적용 
# 3) applymap 은 DataFrame에만 적용
```

```python
# 1) apply() 사용하기 with DataFame

import pandas as pd
df = pd.DataFrame({ 'A': [1,2,3,4], 
                   'B': [10,20,30,40],
                   'C': [20,40,60,80]
                  }, 
                  index=['Row 1', 'Row 2', 'Row 3', 'Row 4'])

df
```

```python
def custom_sum(row):
    return row.sum()
# axis =1(열중심) applying the function to each row

df["D"] = df.apply(custom_sum,axis=1)  
df
```

```python
# axis= 0 (행중심)  applyting the function to each column

df.loc["Row 5"] = df.apply(custom_sum, axis=0)
df
```

```python
# 2) apply() 사용하기 with Series

def multipl_by2(val):
    return val * 2

df["D"]= df["C"].apply(multipl_by2)
df
```

```python
# 3) apply() 사용하기 using lambda 
# 위에 썼던 식을 lambda로 적을 수 있다.

import pandas as pd
df = pd.DataFrame({ 'A': [1,2,3,4], 
                   'B': [10,20,30,40],
                   'C': [20,40,60,80]
                  }, 
                  index=['Row 1', 'Row 2', 'Row 3', 'Row 4'])

df
```

```python
df["D"] = df.apply(lambda x: x.sum(), axis = 1)
df
```

```python
df.loc["Row5"] = df.apply(lambda x: x.sum(), axis = 0)
df
```

```python
df["D"]= df["C"].apply(lambda x: x*2)
df
```

```python
# 1.개별 원소에 함수 Mapping 

import seaborn as sns 
titanic = sns.load_dataset("titanic")
titanic.head(10)
```

```python
# age, fare 열만 가져와보자 
df = titanic.loc[:,["age","fare"]]
df
```

```python
# "ten"이라는 새로운 열을 만들어 10의 값을 넣어보기 
df["ten"] = 10 
df.head(10)
```

```python
# 두 가지 함수 만들기 
def add_10(x):
    return x + 10  # 10을 더해주는 함수 

def add_obj(a,b):
    return a+b  # a, b를 더해주는 함수 

print(add_10(10))
print(add_obj(100,100))
```

```python
# apply 함수를 이용해 시리즈 객체의 개별원소에 적용시켜보자 

# age 열에 적용시켜보자 
df1 = df["age"].apply(add_10)
print(df1.head(10))
```

```python
df2 = df["age"].apply(add_obj,b=10)
print(df2.head(10))
```

```python
df3 = df["age"].apply(lambda x: add_10(x))
df3.head(10)
```

```python
# apply 함수를 이용해 데이터프레임 객체의 개별원소에 적용시켜보자 

df_apply = df.applymap(add_10)
df_apply.head(10)
```

```python
# 시리즈 객체에 함수 매칭 
#  DataFrame객체.apply(매핑함수, axis = ) 함수

# 결측값을 확인하는 함수를 만들어보자 
import seaborn as sns 
titanic = sns.load_dataset("titanic")
df= titanic.loc[:,["age","deck"]]
df.head()
```

```python
def missing_value(x):
    return x.isnull()

result = df.apply(missing_value,axis = 1)
print(result.head())
print('\n')
print(type(result))
```

```python
# 최대값에서 최소값을 뺀 값을 반환하는 사용자함수를 정의

import seaborn as sns 

titanic = sns.load_dataset("titanic")

data = titanic.loc[:,["age","fare"]]


def diff(x):
    return x.max() - x.min()


# 데이터 프레임의 행, 열에대해 적용하기

result = data.apply(diff, axis = 0)
result1 = data.apply(diff, axis = 1)
print(data.max())
print(data.min())
print('\n')
print(data.head())
print('\n')
print(result.head())
print('\n')
print(result1.head())
```

```python
# 함수적용2

def add_two(a,b):
    return a + b

data = data.apply(lambda x: add_two(x["age"],x["fare"]), axis = 1)
data.head()
```
