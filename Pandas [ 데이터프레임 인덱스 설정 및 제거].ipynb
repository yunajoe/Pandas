```python
# set_index: 기존의 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
# reset_index :  기존의 행 인덱스를 제거하고 인덱스를 데이터 열로 추가

import pandas as pd 
import numpy as np

np.random.seed(0)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3, 5), 2)]).T,
                   columns=["C1", "C2", "C3", "C4"])
df1
```

```python
# C1 을 인덱스로 설정해보자

df2 = df1.set_index("C1")
df2
```

```python
# C2 를 인덱스로 설정해보자 
df2.set_index("C2")
```

```python
# reset_index 메서드를 사용하면 일반적인, 보통의 자료열로 바꿀 수 있다. 
```

```python
df2
```

```python
df2.reset_index()
```

```python
# reset_index()를 호출할때 drop = True 로 설정하면 인덱스 열을 일반적인 자료로 만드는것이 아니라 그냥 버리게됨
df2.reset_index(drop = True)  # index였던 C1이 사라짐
```

```python
# 다중인덱스
# column 를 생성할 때 리스트 형태로 넣으면 다중 인덱스를 갖게 된다 
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C1", "C2", "C1", "C2"]])
df3
```

```python
# 다중 인덱스는 이름을 지정하여 넣으면 더 편리하다 
df3.columns.names = ["인덱스1","인덱스2"]
df3
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"],
                          ["id_" + str(i + 1) for i in range(3)] * 2])
df4.columns.names = ["Cidx1", "Cidx2"]
df4.index.names = ["Ridx1", "Ridx2"]
df4
```

```python
# 행 인덱스와 열 인덱스 교환
# stack >> 열 인덱스를 행 인덱스로 변환 
# unstack >> 행 인덱스를 열 인덱스로 변환

df4.stack("Cidx1")
```

```python
df4.stack("Cidx2")
```

```python
df4
```

```python
df4.unstack("Ridx1")
```

```python
df4.unstack("Ridx2")
```
